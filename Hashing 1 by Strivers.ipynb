```python

n = int(input()) # size of array

a = list() # empty list
# a = list(map(int,input().split()))

for i in range(n):
    a.append(int(input()))

t = int(input()) # number of queries

############## pre compute
h = [0]*20 #[0,0,0,0,0,..... 0] 20 times

for i in range(n):
    h[a[i]] += 1

print(h)

while(t>=0):
    num = int(input())
    print(h[num])
    t = t - 1
```

```python
 ############ for small alphabets   @@@@!!!!!!!!!!!!


s = input()          ## string .... imagine all are small alphazets
t = int(input())     # number of queries

############## pre compute

h = [0]*26            #[0,0,0,0,0,..... 0] 26 times

for i in s:
    h[ord(i) - 97] += 1

print(h)

while(t >= 0):
    c = input()
    print(h[ord(c) - 97])           ########### fetching
    t = t - 1
```

```python
###########    for all characters    ############


s = input()          ## string .... imagine all are small alphazets
t = int(input())     # number of queries

############## pre compute

h = [0]*256           #[0,0,0,0,0,..... 0] 26 times

for i in s:
    h[ord(i)] += 1

#print(h)

while(t >= 0):
    c = input()
    print(h[ord(c)])           ########### fetching
    t -= 1
```
